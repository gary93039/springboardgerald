```python
import pandas as pd
import json
from pandas.io.json import json_normalize

json.load((open('data/world_bank_projects.json')))
worlddf=pd.read_json('data/world_bank_projects.json')
worlddf.head()
```

```python
#finding the top 10 major project themes
data=json.load((open('data/world_bank_projects.json')))
normalize_df= json_normalize(data,'mjtheme_namecode',['_id','countryname'])
```

```python
csort_df =normalize_df.set_index(['code']).sort_index()

csort_df
```

```python
csort_df['name'] =csort_df.name.replace('',method ='ffill')

csort_df=csort_df.reset_index()     
csort_df['_id'] = csort_df['_id'].astype(str)  
csort_df=csort_df.drop_duplicates(['_id','code'], keep='first')  

q2_df = csort_df.groupby(['code','name']).size().sort_values(ascending=False).head(10)#group df by size of code and sort by descending

q2_df
```
